```python
import sys
sys.path.append('../')
import pandas as pd
import seaborn as sns; sns.set(style='ticks', context='paper',
                               rc={'font.size': 6, 
                                   'axes.titlesize': 7,
                                   'axes.labelsize': 7,
                                   'axes.labelpad': 2,
                                   'xtick.labelsize': 5,
                                   'ytick.labelsize': 5,
                                   'legend.fontsize' : 5.5,
                                   'legend.title_fontsize' : 7,
                                   'legend.markerscale': 0.5,
                                   'axes.spines.top': False,
                                   'axes.spines.right': False,
                                   'axes.linewidth': 0.4,
                                   'xtick.major.width': 0.4,
                                   'ytick.major.width': 0.4,
                                   'xtick.major.size': 2.5,
                                   'ytick.major.size': 2.5,
                                   'xtick.minor.size': 1.5,
                                   'ytick.minor.size': 1.5,
                                   'xtick.minor.width': 0.2,
                                   'ytick.minor.width': 0.2,
                                   'figure.constrained_layout.use': True,
                                   'figure.dpi': 200})
from matplotlib import pyplot as plt
from gaussian_spheres import run_simulation
```

```python
method = 'proj'

data_default = run_simulation(method=method, n_repeats=100)

data_eco = pd.DataFrame()
for dim_eco in [3, 5, 10, 20]:
    data_eco = data_eco.append(run_simulation(dim_eco=dim_eco, method=method, n_repeats=100))

data_exp = pd.DataFrame()
for dim_exp in [1, 3, 5, 10]:
    data_exp = data_exp.append(run_simulation(dim_exp=dim_exp, method=method, n_repeats=100))

data_shared = pd.DataFrame()
for max_shared in [1, 3, 5, 10]:
    data_shared = data_shared.append(run_simulation(max_shared=max_shared, method=method, n_repeats=100).assign(max_shared=max_shared))
```

```python
fig, axs = plt.subplots(2, 2, figsize=(4.5, 3.5))

sns.lineplot(data=data_default, x='dim_model', y='r2', 
             ci='sd', ax=axs[0, 0])

sns.lineplot(data=data_eco, x='dim_model', y='r2', hue='dim_eco',
             ci='sd', palette=sns.color_palette('crest')[:4], ax=axs[0, 1])
axs[0, 1].legend(loc='lower right')

sns.lineplot(data=data_exp, x='dim_model', y='r2', hue='dim_exp',
             ci='sd', palette=sns.color_palette('crest')[:4], ax=axs[1, 0])
axs[1, 0].legend(loc='lower right')

sns.lineplot(data=data_shared, x='dim_model', y='r2', hue='max_shared',
             ci='sd', palette=sns.color_palette('crest')[:4], ax=axs[1, 1])
axs[1, 1].legend(loc='lower right')

axs[0, 0].set(title='Example', xlabel=r'$D_{model}$', ylabel=r'Encoding score ($R^2$)', xlim=(0, 21), ylim=(-0.1, 1.1), xticks=range(1, 20, 4))
axs[0, 1].set(title=r'Varying $D_{ecological}$', xlabel=r'$D_{model}$', ylabel=None, xlim=(0, 21), ylim=(-0.1, 1.1), xticks=range(1, 20, 4))
axs[1, 0].set(title=r'Varying $D_{experimental}$', xlabel=r'$D_{model}$', ylabel=r'Encoding score ($R^2$)', xlim=(0, 21), ylim=(-0.1, 1.1), xticks=range(1, 20, 4))
axs[1, 1].set(title=r'Varying $D_{ecological}$ captured by model', xlabel=r'$D_{model}$', ylabel=None, xlim=(0, 21), ylim=(-0.1, 1.1), xticks=range(1, 20, 4))
axs[0, 1].axes.yaxis.set_ticklabels([])
axs[1, 1].axes.yaxis.set_ticklabels([])

plt.savefig('/home/eric/Downloads/simulation_projections.jpg', dpi=300)
plt.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(4.5, 2))

sns.lineplot(data=data_default, x='dim_model', y='r2', 
             ci='sd', ax=axs[0])

sns.lineplot(data=data_eco, x='dim_model', y='r2', hue='dim_eco',
             ci='sd', palette=sns.color_palette('crest')[:4], ax=axs[1])
axs[1].legend(loc='lower right', title=r'$D_{ecological}$')

axs[0].set(title=r'$D_{ecological} = 20,\; D_{experimental} = 3$', xlabel=r'$D_{model}$', ylabel=r'Encoding score ($R^2$)', xlim=(0, 21), ylim=(-0.1, 1.1), xticks=range(1, 20, 4))
axs[1].set(title=r'Varying $D_{ecological}$', xlabel=r'$D_{model}$', ylabel=None, xlim=(0, 21), ylim=(-0.1, 1.1), xticks=range(1, 20, 4))
axs[1].axes.yaxis.set_ticklabels([])

plt.savefig('/home/eric/Downloads/simulation_projections.jpg', dpi=300)
plt.show()
```
